```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import random
```

```python

def funcao1(x):
  return (1-x*x)

def funcao2(x):
  return np.exp(x)

def funcao3(x):
  return np.sin(x)**2

def mc_it_img(N_total,a,b,f_max,type_it):
  N_hit = np.float64(1.0)

  x_in, x_out = [], []
  y_in, y_out = [], []

  for i in range(N_total):
    x = random.uniform(a,b)
    y = random.uniform(0,f_max)

    #print(x,y)

    if(type_it == 1):
      fx = funcao1(x)
    elif(type_it == 2):
      fx = funcao2(x)
    elif(type_it == 3):
      fx = funcao3(x)
    else:
      fx = -1

    if y < fx:
      N_hit += 1
      x_in.append(x)
      y_in.append(y)
    else:
      x_out.append(x)
      y_out.append(y)

  return ((a+b)*f_max)*(N_hit/N_total), x_in, x_out, y_in, y_out

def mc_it_hist(N_total,a,b,f_max,type_it):
  
  ans_array = np.zeros(10000)

  for j in range(10000):
    N_hit = np.float64(1.0)

    for i in range(N_total):
      x = random.uniform(a,b)
      y = random.uniform(0,f_max)

      if(type_it == 1):
        fx = funcao1(x)
      elif(type_it == 2):
        fx = funcao2(x)
      elif(type_it == 3):
        fx = funcao3(x)
      else:
        fx = -1

      if y < fx:
        N_hit += 1

    ans_array[j] = ((a+b)*f_max)*(N_hit/N_total)
    # print(ans_array[j])
  
  # print(ans_array)
  return ans_array
```

## Primeira integral

```python
v_est, x_in, x_out, y_in, y_out = mc_it_img(100,0,1,1,1)
xlabel = np.linspace(0,1,1000)
plt.plot(xlabel, 1-xlabel**2,'k')
plt.plot(xlabel, np.ones(len(xlabel)),'k')
plt.plot(np.ones(len(xlabel)), xlabel,'k')
plt.plot(xlabel, np.zeros(len(xlabel)), 'k')
plt.plot(np.zeros(len(xlabel)), xlabel, 'k')
plt.scatter(x_in, y_in, marker='.')
plt.scatter(x_out, y_out, marker='.')
```
