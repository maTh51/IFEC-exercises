```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
def population_dynamics(n,r,x0):
    if(n == 0):
        xn = x0
    else:
        xn = population_dynamics(n-1,r,x0)
        
    return r*xn*(1-xn)
```

4. Considere 10000 valores de r igualmente espaçados entre 10−5 e 4. Para cada um desses valores, vamos iterar a equação logística 1000 vezes, i.e., nmax = 1000 e plotar num gráfico os últimos 100 passos para cada valor de r. A figura gerada é conhecida como diagrama de bifurcação. Essa figura trás inúmeras informações importantes. Tire um tempo para pensar nela e analisar o que encontrou

```python
nmax = 1000
r_vec = np.linspace(3.840,3.856,10000)
y_global = [] # array de arrays
x0 = 0.5

pd = {}

def population_dynamics(n,r,x0):
    if n in pd.keys():
        return pd[n]

    if(n == 0):
        xn = x0
    else:
        xn = population_dynamics(n-1,r,x0)
    
    pop = r*xn*(1-xn)
    pd[n] = pop
    return pop

for r in r_vec:
    y_vec = []
    for n in range(0, nmax):
        value = population_dynamics(n,r,x0)
        if n >= nmax-100:
            y_vec.append(value)

    y_global.append(y_vec)
    pd.clear()

#print(y_global[0])
```

```python
plt.plot(r_vec,y_global)
ax = plt.gca()
ax.set_ylim([0.44, 0.56])
plt.grid()
plt.show()
```
