```python
def uni_walk(n:int):
  walks = [];
  # initial walk
  walks.append(np.random.rand() - 0.5);
  for i in range(1,n):
    walks.append(np.random.rand() - 0.5 + walks[i-1])

  return walks
```

```python
uni_walk(10)
```

```python
def bi_walk(n: int):
  walks = []
  random_value1 = np.random.rand() - 0.5
  random_value2 = np.random.rand() - 0.5
  # initial bi walk
  walks.append([random_value1, random_value2]);
  for i in range(1, n):
    new_rand1 = np.random.rand() - 0.5 + walks[i-1][0]
    new_rand2 = np.random.rand() - 0.5 + walks[i-1][1]
    walks.append([new_rand1, new_rand2])
  return walks
```

```python
bi_walk(10)
```

```python
# n ew routine for the last walks in 2d
def new_bi_walk(w:int, n:int):
  walks = []
  for i in range(w):
    new_walk = bi_walk(n)
    x_values = []
    y_values = []
    for value in new_walk:
      x_values.append(value[0])
      y_values.append(value[1])

    walks.append([x_values[-1], y_values[-1]])

  return walks
```

```python
# generate new routine walks for 1 and 10
final_walk_1 = new_bi_walk(10000, 1);
final_walk_10 = new_bi_walk(10000, 10)
```

```python
# generate a scatter plot of the two new routines in the same graph 
x_final_1_values = []
y_final_1_values = []
x_final_10_values = []
y_final_10_values = []

# extract the x and y values for both steps
for value in final_walk_1:
  x_final_1_values.append(value[0])
  y_final_1_values.append(value[1])

for value in final_walk_10:
  x_final_10_values.append(value[0])
  y_final_10_values.append(value[1])


plt.scatter(x_final_10_values, y_final_10_values)
plt.scatter(x_final_1_values, y_final_1_values)
plt.title("Scatter plot of the finals values of the walk")
plt.xlabel('X')
plt.ylabel('Y')

```

```python
# plotting of histogram 

def last_uni_walks(w:int, n:int):
  walks = []
  for i in range(w):
    new_walk = uni_walk(n)
    walks.append(new_walk[-1])

  return walks
```

```python
# calcualte the gaussian value 
def gaussian(n:int):
  a = -0.5
  b = 0.5
  variance = ((b-a)**2)/12
  sigma = np.sqrt(n) * np.sqrt(variance)
  x = np.linspace(-3*sigma, 3*sigma, 100);
  plt.plot(x, norm.pdf(x, 0, sigma))
```

```python
# plot histogram 
def historgram(w:int, n:int, bins:int, title:str):
  last_walk = last_uni_walks(w, n);
  plt.hist(last_walk, bins=bins, density=True)
  gaussian(n)
  plt.title(title)
```

```python
# final unidimensional walk with 1 step
historgram(10000, 1, 50, "histogram with one step")
```
